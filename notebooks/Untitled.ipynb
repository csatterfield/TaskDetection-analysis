```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel("mr.xlsx", sheet_name="Sheet1")
```

```python
my_pal = {"Respondent": "#FFAD05", "Developer": "#5595ED"}
sns.set(style="whitegrid")
ax = sns.violinplot(x=df["Type"], y=df["Accuracy"], palette=my_pal, saturation=1)
ax.set_xlabel('')
ax.set(ylim=(0,1.0))
ax.set(yticks=np.arange(0,1.1,0.1))
plt.title("Accuracy Distributions - Per Respondent and Per Developer")
plt.savefig("fig.pdf", bbox_inches="tight")
```
